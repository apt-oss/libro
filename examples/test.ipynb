```
from tqdm import tqdm
import time
for epoch in tqdm(range(100), desc="Epoch", position=0):
    time.sleep(0.1)
```

```
import time
for i in range (0, 500): 
    epoch = i
    batch_idx = i % 100
    time.sleep(0.1)
    print('Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: 2.797897966'.format(
        epoch, batch_idx * 100, 1000,
        100. * batch_idx / 1000))
```
